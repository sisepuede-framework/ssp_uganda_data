## Source

**Private and Public Mode Shares:**

* These variables are based on Table 4-1, page 42.
* We sum the sample counts for the following mappings:

  * Car → `road_light`
  * Station Wagon → `road_light`
  * Pickups → `road_light`
  * Minibus → `public`
  * Medium Bus → `public`
  * Large Bus → `public`
  * Motorcycles → `powered_bikes`
* After calculating the totals, we compute the mode shares using the corresponding categories.

**Regional Mode Shares:**

* Calculated based on transport demand data from `PASSENGER_TRANSPORT_DEMAND.ipynb`.

```python
# Set up paths
CURRENT_DIR_PATH = os.getcwd()
OUTPUT_DATA_DIR_PATH = os.path.join(CURRENT_DIR_PATH, "output_data")
INPUT_DATA_DIR_PATH = os.path.join(CURRENT_DIR_PATH, "input_data")
```

```python
base_year = 2019

# human_powered``|``powered_bikes``|``public``|``road_light``|``water_borne
private_and_public_mode_shares = {
    "year": base_year,
    "frac_trns_pkm_dem_private_and_public_human_powered": 0.0,
    "frac_trns_pkm_dem_private_and_public_powered_bikes": 0.5547,
    "frac_trns_pkm_dem_private_and_public_public": 0.1260,
    "frac_trns_pkm_dem_private_and_public_road_light": 0.3193,
    "frac_trns_pkm_dem_private_and_public_water_borne": 0.0
}


# aviation``|``rail_passenger``|``road_heavy_regional``|``road_light``|``water_borne
regional_mode_shares = {
    "year": base_year,
    "frac_trns_pkm_dem_regional_aviation": 0.9918,
    "frac_trns_pkm_dem_regional_rail_passenger": 0.0002,
    "frac_trns_pkm_dem_regional_road_heavy_regional": 0.0,
    "frac_trns_pkm_dem_regional_road_light": 0.0,
    "frac_trns_pkm_dem_regional_water_borne": 0.0080
}
```

```python
# Convert to DataFrames
df_private_public = pd.DataFrame([private_and_public_mode_shares])
df_regional = pd.DataFrame([regional_mode_shares])
```
