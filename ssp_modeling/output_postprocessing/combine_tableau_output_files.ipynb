```python
import pandas as pd
import os
```

```python
SCRIPT_DIR_PATH = os.getcwd()
PARENT_DIR_PATH = os.path.dirname(SCRIPT_DIR_PATH)
TABLEAU_OUTPUT_DIR = os.path.join(PARENT_DIR_PATH, 'Tableau', 'data')
```

```python
df_period_5 = pd.read_csv(os.path.join(TABLEAU_OUTPUT_DIR, 'emissions_uganda_WIDE_INPUTS_OUTPUTS_period_5.csv'))
df_period_5
```

```python
df_period_7 = pd.read_csv(os.path.join(TABLEAU_OUTPUT_DIR, 'emissions_uganda_WIDE_INPUTS_OUTPUTS_period_7.csv'))
df_period_7
```

```python
df_period_7.strategy.unique()
```

```python
# Filter df_period_7 to only include rows with strategy values 'Strategy TX:BASE', 'NZ', and 'Historical'
df_period_7_filtered = df_period_7[df_period_7['strategy'].isin(['Strategy TX:BASE', 'NZ', 'Historical'])]
df_period_7_filtered
```

```python
# Filter df_period_5 to only include rows with strategy with value 'NDC_2'
df_period_5_filtered = df_period_5[df_period_5['strategy'] == 'NDC_2']
df_period_5_filtered
```

```python
# Concat the filtered DataFrames
df_combined = pd.concat([df_period_7_filtered, df_period_5_filtered], ignore_index=True)
df_combined
```

```python
print(df_combined.shape)
print(df_period_5.shape)
print(df_period_7.shape)
```

```python
df_combined.strategy.unique()
```

```python
df_combined.to_csv(os.path.join(TABLEAU_OUTPUT_DIR, 'emissions_uganda_WIDE_INPUTS_OUTPUTS_combined.csv'), index=False)
```
